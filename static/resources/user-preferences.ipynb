## ShareMyPlaylists Scraping

```python
"""
Function
--------
get_top_members

Scrapes top 100 members from playlists.net (http://playlists.net/chart/top-100-members)
and returns a list of member names

Parameters
-----------
None

Returns
--------
members : list
    List of strings, each corresponding to a top 100 member
"""

def get_top_members():
    url = 'http://playlists.net/chart/top-100-members'
    html = requests.get(url).text
    parsed_html = BeautifulSoup(html)
    members_list = parsed_html.body.findAll("a", attrs={'class':'item'})
    members = [str(member['href']) for member in members_list]
    return members
```

```python
#get list of top 100 members
top_members_data = get_top_members()
```

```python
"""
Function
--------
get_playlists

Scrapes all the playlists created by a member
and returns a list of playlist names

Parameters
-----------
member: string 
    String from list of member names (top_members_data)

Returns
--------
playlists : list
    List of strings, each corresponding to a playlist created by the member
"""

def get_playlists(member):
    url = 'http://playlists.net/' + member + '/playlists'
    html = requests.get(url).text
    parsed_html = BeautifulSoup(html)
    
    #find number of pages used to list all of user's playlists
    paging = parsed_html("div", attrs={'class': 'paging clear-fix'})[0]
    pages = str(paging.p.contents[0])
    page_last = int(pages.split(' ')[-1])

    playlists = []
    #for each page of the user's playlists, scrape playlist names
    for i in xrange(1, page_last+1):
        url2 = 'http://playlists.net/' + member + '/playlists/index/page/%i' % i
        html2 = requests.get(url2).text
        parsed_html2 = BeautifulSoup(html2)
        playlists_list = parsed_html2.body.findAll("li", attrs={'class':'clear-fix playlist'})

        for plist in playlists_list:
            playlists.append(str(plist.a['href']))

    return playlists
```

```python
#get playlists created by top 100 members and combine into one list
playlists_data = []
for m in top_members_data:
    playlists_data.extend(get_playlists(m))
```

```python
"""
Function
--------
get_songs

Scrapes all the songs in a given playlist
and returns a list of song names

Parameters
-----------
plist: string 
    String from list of playlist names for a member

Returns
--------
songs : list
    List of tuples, each corresponding to a song and its artist
"""

def get_songs(plist):
    url = 'http://playlists.net/' + plist
    html = requests.get(url).text
    parsed_html = BeautifulSoup(html)
    iframes = parsed_html.body.findAll("iframe")

    #check for url of embedded spotify playlist
    if len(iframes) == 2:
        #scrape url
        url2 = str(iframes[1]['src'])
        html2 = requests.get(url2).text
        parsed_html2 = BeautifulSoup(html2)
        songs_list = parsed_html2.body.findAll("li", attrs={'rel':'track'})

        songs = []    
        for song in songs_list:
            info = song.ul.findAll("li") #list in which second item corresponds to song, third item corresponds to artist
            
            #remove enumeration for song name
            song_enum = str(info[1].contents[0])
            song_name_parts = song_enum.split('.')
            
            song_name = ''
            #account for song names with '.'
            song_name = song_name.join(song_name_parts[1:])
            song_name = song_name[1:]
            
            #append tuple of (song name, artist)
            songs.append((song_name, str(info[2].contents[0])))    
        return songs
    else:
        #if no url, return empty list
        return []
```

```python
#this takes a few hours

#build dict, with keys as playlist names and values as songs
songs_data = {}
for p in playlists_data:
    songs = get_songs(p)
    songs_data[p] = songs
```

```python
"""
Function
--------
playlists_int

Takes a dict, with keys as playlist names and values as songs (like songs_data2),
and a set of songs of interest,
and returns a dict, with keys as playlist names and values as songs from the playlists that are also found in the set of songs of interest

Parameters
-----------
playlists_dict: dict 
    Dict, with keys as playlist names and values as set of songs
    
songs_set: set
    Set of songs of interest

Returns
--------
playlists_set_int: dict
    Dict, with keys as playlist names and values as songs found in both original playlist and set of songs
"""

def playslists_int(playlists_dict, songs_set):
    #new dict to build
    playlists_set_int = {}
    
    for plist in playlists_dict.keys():
        plist_set = playlists_dict[plist]
        plist_set_int = plist_set.intersection(songs_set)
        
        #add pared down set of songs to new dict
        playlists_set_int[plist] = plist_set_int
    
    return playlists_set_int
```

```python
"""
Function
--------
song_graph

Takes a list of playlists
and returns a NetworkX graph with nodes as songs and edges representing co-occurrence of two songs
on a playlist

Parameters
-----------
data: list of lists 
    List of playlists, each of which will be used to build song connections

Returns
--------
G: NetworkX graph
    Graph with nodes as songs and edges representing co-occurrences (weighted by number of co-occurrences)
"""

from itertools import combinations
import networkx as nx

def song_graph(data):
    G = nx.Graph()
    #iterate through data
    #use itertools combinations to find pairs of song TUPLES
    for plist in data:
        for (a,b) in combinations(plist, 2):
            song1 = a
            song2 = b
            if G.has_edge(song1, song2):
                w = G[song1][song2]['weight']
                w += 1
                G.add_edge(song1, song2, weight=w, difference=1/float(w))
            else:
                G.add_edge(song1, song2, weight=1, difference=1.)

    return G
```

```python
#determine song interactions in songs_data_int
graph_songs_top = song_graph(songs_data_int.values())
```

### 2) Analyze closeness centrality
We decide to use closeness centrality to identify nodes that are most closely connected to other nodes the graph. Closeness centrality for a node is determined by the inverse of the sum of its distances to all other nodes. 

```python
songs_centrality = nx.closeness_centrality(graph_songs_top, distance="difference")
songs_centrality = songs_centrality.items()

songs_sorted = sorted(songs_centrality, key=itemgetter(1))

print "Lowest centralities:", songs_sorted[:5]
print "Highest centralities:", songs_sorted[-5:]

centralities = map(itemgetter(1), songs_sorted)
xlocs = range(len(centralities))
plt.bar(xlocs, centralities)

plt.xlim(min(xlocs), max(xlocs))
plt.ylim(0, max(centralities))
plt.ylabel('Centrality')
plt.xlabel('Songs')
plt.title('Closeness Centralities for Songs')

remove_border()
```

**Observations:** The songs with the highest centralities are not necessarily the songs that occur on the most number of playlists, though there is some positive correlation. While 'Smells Like Teen Spirit', which has the highest number of occurrences, is also the song with the highest centrality, the other songs with high centrality include a number of Guns N' Roses songs that don't necessarily have the highest occurrence. However, it is likely that these songs have been listed together a number of times on the same playlists, which gives their mutual edges a larger weight/smaller distance and, overall, a larger centrality.

```python
from scipy import stats

attributes = ['occ', 'cent', 'score', 'year']
for a, b in combinations(attributes, 2):
    #used Spearman because data are not normal
    r, p = stats.spearmanr(songs_top276_attr_small[a], songs_top276_attr_small[b])
    print a, b, 'p-value:', p
```

**Observations:** We performed a Spearman correlation, as opposed to a Pearson correlation, because the data are not normal.

As expected, there is a high correlation between occurrences of songs on different playlists and closeness centrality of each song. This is because songs that appear on different playlists with different songs will therefore have smaller distance to a variety of songs in the network.

Interestingly, there appears to be a correlation between centrality and year. It seems that songs that are less recent tend to have a higher centrality value. Understandably, there are many more songs in the data set that are recent rather than old, therefore there is more opportunity for the recent songs to have a range of centralities. In addition, it is important to note that our data set has a size of 276, which is still relatively small and renders the p-value slightly less reliable.

However, it is also reasonable to consider that older songs that are less appreciated will not persist in many playlists; only the timeless classics will remain and be highly represented in current-day playlists. On the other hand, newer songs have yet to be tried in various playlists and have not yet had the time to rise to prominence above their peers (i.e. be featured in more playlists than usual).

We saw a similar phenomenon in HW3, where older movies tended to have higher ratings, whereas those for newer movies were across the board. This observation constitutes a form of selection bias, by which only old (i.e. released before the use of online movie databases) movies that were good are worth being memorialized in a movie database, whereas all new movies have the chance to be criticized/praised on the database. The platform of online playlists acts similarly, memorializing the songs that are deemed classics.

```python
"""
Function
--------
graph_nodes_attr

Takes a NetworkX graph and a DataFrame with attributes to be added to each of the nodes
and returns a NetworkX graph with atributes added to each node

Parameters
-----------
G: NetworkX graph, such as graph_songs_top
    Graph with nodes as songs and edges representing co-occurrences (weighted by number of co-occurrences)
    
attr: DataFrame, such as songs_top276_attr_small
    DataFrame lists each song/artist with occ (occurrences), mod (modularity), score, year, and genre

Returns
--------
G2: NetworkX graph
    Graph, with attributes from attr added to G
    
"""

def graph_nodes_attr(G, attr):
    #create a copy of G
    G2 = G.copy()
    
    #iterate through each row of attr DataFrame
    for index, row in attr.iterrows():
        node = (row['name'], row['artist'])
        
        #add attributes to each node
        G2.node[node]['occ'] = int(row['occ'])
        G2.node[node]['cent'] = float(row['cent'])
        G2.node[node]['mod'] = int(row['mod'])
        G2.node[node]['score'] = int(row['score'])
        G2.node[node]['year'] = int(row['year'])
        G2.node[node]['genre'] = row['genre']
    
    return G2
```

```python
#add attributes from DataFrame songs_top276_attr_small to NetworkX Graph graph_songs_top
graph_top_attr = graph_nodes_attr(graph_songs_top, songs_top276_attr_small)
```
