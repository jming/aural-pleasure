```python
%matplotlib inline

import json

import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import networkx as nx
import itertools

from IPython.display import Image

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

And in order to get all of the information for each of the artists, we wrote functions to get the artist id and the artist events.

```python
# function to get the Songkick API artist id based on artist name text
def get_artistid(artist):
    
    # set up URL and options
    url = 'http://api.songkick.com/api/3.0/search/artists.json'
    options = {'apikey': api_key, 'query': artist}
    
    # get JSON data
    data = requests.get(url, params=options).text
    data = json.loads(data)
    
    # if no data, return None
    if data['resultsPage']['totalEntries'] == 0:
        return None
    
    # return the artist id
    return str(data['resultsPage']['results']['artist'][0]['id'])

lorde = get_artistid('lorde')
print lorde
```

```python
# function to get a list of all events an artist has performed at based on artist id
def get_artist_events(artistid):
    
    # set up the url
    total_pages = 1
    i = 0
    url = 'http://api.songkick.com/api/3.0/artists/%s/gigography.json' % artistid
    artist_events = []
    
    # loop through the different pages of the request result
    while i <= total_pages:
        
        i += 1
        
        # get the data
        options = {'apikey': api_key, 'page': i}
        data = requests.get(url, params=options).text
        data = json.loads(data)
        
        total_entries = int(data['resultsPage']['totalEntries'])
        
        # figure out total pages
        if i == 1:
            total_pages = total_entries / 50
            if total_entries % 50 == 0:
                total_pages -= 1
        
        # make sure there are entries
        if total_entries == 0:
            return None
        
        # append all of the ids of the events to artist_events
        for event in data['resultsPage']['results']['event']:
            artist_events.append(event['id'])
    
    # return list of events
    return artist_events

lorde_events = get_artist_events(lorde)
print lorde_events
```

Then we combined the two functions to look at getting a list of events for all of the artists of a given list of artists and use this dictionary of values to create a graph.

```python
# get all the artist events for a list of artists
def get_all_artist_events(artists):
    
    all_artist_events = {}
    
    # loop through artists
    for artist in artists:

        # get artist id
        artistid = get_artistid(artist)
        
        # get artist events (if artist exists)
        if artistid:
            artist_events = get_artist_events(artistid)
            if artist_events:
                all_artist_events[artist] = artist_events

    # return dictionary of artist:[events]
    return all_artist_events
```

```python
# create an artist graph based on a dictionary of artists mapped to a list of events
def artist_graph(all_artist_events):
    
    graph = nx.Graph()
    
    # loop through each combination of artists
    for (artist1, artist2) in itertools.combinations(all_artist_events.keys(),2):
        
        # create nodes for each of the artists
        graph.add_node(artist1)
        graph.add_node(artist2)
        
        # add an edge if there are events in common
        events_in_common = len(set(all_artist_events[artist1]) & set(all_artist_events[artist2]))
        if events_in_common > 0:
            graph.add_edge(artist1, artist2, weight=events_in_common)
        
    # calculate the difference of each of the edges
    for (a1,a2) in graph.edges():
        data = graph.get_edge_data(a1,a2)
        graph.add_edge(a1, a2, weight=data['weight'], difference=1./data['weight'])
        
    # return completed graph
    return graph
```

```python
# function to display a basic networkx graph
def show_graph(graph):
    
    # make sure draw_spring results same at each call
    np.random.seed(1)
    
    # determine pos of each node using a spring layout
    pos = nx.spring_layout(graph, iterations=200)
    
    # plot edges
    nx.draw_networkx_edges(graph, pos, alpha=.05)
    
    # plot the nodes
    nx.draw_networkx_nodes(graph, pos, node_color='red')
    
    # draw the labels
    lbls = nx.draw_networkx_labels(graph, pos, alpha=5, font_size=8)
    
    # get rid of coordinates
    plt.xticks([])
    plt.yticks([])
    remove_border(left=False, bottom=False)
```

## Visualizing the networks using D3

Due to the limited ability to visually analyze the networks using Networkx and Gephi, we sought out a way to display the information interactively, especially with the website, or final product in mind. We turned to D3.js, a Javascript-based library used for interactive visualizations that could be hosted on the web.

We used the force-directed graph element of D3 and based it off of an example of the Les Miserables character co-occurence from http://bl.ocks.org/mbostock/4062045.

In order to do this, we used jQuery to parse through the 'playlists_artist_graph.gexf'that was created as an XML file to get the information for each of the nodes and links. 

```python
# read in the data for the artist graph with info
d3.xml("../static/resources/artist_graph_withinfo.gexf", "application/xml", function(gexf) {

  # get the nodes from the xml file
  var xml_nodes = gexf.documentElement.getElementsByTagName('node');
  var node_names = new Array();
  var nodes = new Array();
  
  # extract information from each xml node
  for (n = 0; n< xml_nodes.length; n++) {
    node = xml_nodes[n];
    attrs = node.getElementsByTagName('attvalue')
    nodes.push({
      name: node.id,
      genre: attrs[0].getAttribute('value'),
      pagerank: attrs[1].getAttribute('value'),
      modularity: attrs[2].getAttribute('value'),
      score: attrs[3].getAttribute('value')
    });
    node_names.push(node.id);
  }

  # get links from xml file
  var links = new Array();
  var xml_links = gexf.documentElement.getElementsByTagName('edge');

  # extract information from each xml link
  for (l=0; l < xml_links.length-1; l++) {
    link = xml_links[l];
    var source_name = link.getAttribute('source');
    var target_name = link.getAttribute('target');
    links.push({
      source: node_names.indexOf(source_name),
      target: node_names.indexOf(target_name),
      value: link.getAttribute('weight')
    })
  }
  # continue to build the graph here
});
```

**Note.** We later changed this scraping to convert CSV and .gexf files into JSON, which would save processing power when loading the webpage--latency was a huge problem we were facing. The code for convering the CSV and .gexf files into JSON are included below.

```python
# function to take in df of nodes and edge info into json

def df_to_json(nodes,edges):
    
    results = {'nodes': [], 'edges': []}
    names = []
    
    # get node information
    for index,node in nodes.iterrows():
        results['nodes'].append({'name':node['Id'], 'genre':node['genre'], 'score':node['score'], 'modularity': node['Modularity Class'], 'pagerank': node['PageRank']})
        names.append(node['Id'])
        
    # get edge information
    for index,edge in edges.iterrows():
        results['edges'].append({'source':names.index(edge['Source']), 'target':names.index(edge['Target']), 'value':edge['Weight']})

    return results
```

## Improving D3 visualization with more attributes

Then, in order to include more information in each of the nodes, we also included the genre and score from previous scraping of information as well as the page rank and modularity as calculated by Gephi and exported as a CSV.

To do this, we downloaded the nodes and edges information from Gephi as a CSV.

```python
# get previous song data

all_songs = pd.read_csv('songs_top276_attr_small.csv')
artist_graph_nodes = pd.read_csv('playlists_artist_graph_nodes.csv')
artist_graph_edges = pd.read_csv('playlists_artist_graph_edges.csv')

all_songs_byartist = all_songs.groupby('artist')
```

And then we combined this with the more detailed information in songs_top276_attr_small.csv. In order to determine the 'genre' of an artist, we returned the most common element of the list of genres of songs an artist has performed. 

```python
# returns the most common element in a list or the first element

def most_common(lst):
    return max(((item, lst.count(item)) for item in set(lst)), key=lambda a: a[1])[0]
```

```python
# get more information for each of the artists

def get_artist_scores_genres(artists):
    
    genres = []
    scores = []
    
    # loop through artists and append genres and scores
    for artist in artists:
        if artist == 'Bonobo':
            genres.append('no-genre')
            scores.append(0)
        else:
            songs = all_songs_byartist.get_group(artist)
            genres.append(most_common(songs['genre'].tolist()))
            scores.append(sum(songs['score'].tolist()))

    return pd.DataFrame.from_dict({'Id': artists, 'genre': genres, 'score': scores})
```

```python
# get list of artists based on graph nodes

artist_list = artist_graph_nodes['Id'].tolist()
artist_scores_genres = get_artist_scores_genres(artist_list)
```

```python
# create merged dataframe of new data and old data

artist_withinfo = artist_scores_genres.merge(artist_graph_nodes, on='Id')
print artist_withinfo.head()
```

```python
# create an artist graph with all of the new information

def artist_graph_withinfo(nodes, edges):
    
    graph = nx.Graph()
    
    for index,node in nodes.iterrows():
        graph.add_node(node['Id'], genre=node['genre'], score=node['score'], modularity=node['Modularity Class'], pagerank=node['PageRank'])
    
    for index,edge in edges.iterrows():
        graph.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'], difference=edge['difference'])
    
    return graph

artist_graph_withinfo = artist_graph_withinfo(artist_withinfo, artist_graph_edges)
```

```python
# write this into .gexf format

nx.write_gexf(artist_graph_withinfo, 'artist_graph_withinfo.gexf')
```

However, it still looks like a massive blob. We tried adjusting a couple of parameters in the force directed graph in order to make the visualization easier to deal with. Some of the parameters include:

 * **charge** is how the nodes interact with each other - a positive charge results in node attraction and a negative charge results in node repulsion. The default value is -30, and we found we would have to substantially decrease the amount into larger negative numbers to spread the nodes out from each other. However, a charge that is too negative results in unweidly graphs that often propel itself off the page.


 * **gravity** is the gravitational strength toward the center of the page, as if it were a spring. We found that it was necessary to have strong gravity to make sure the graph does not float off the screen because its own nodes are repelling each other, but the strong gravity must be balanced with charge to make sure the nodes do not scrunch up again.
 

 * **friction** determines how the nodes move around on the page. Lower friction results in more fluid ability and too much motion but too high of a friction paramter would result in nodes that are too static.
 
 
 * **linkDistance** determines how far apart two nodes would be, with a default of 20. We felt that the link distance should be a function of the nodes, so we found a function that would adequately fit our formatting needs.
 
 
 * **linkStrength** is the rigidity of the links. We found that links that are more rigid end up propelling their nodes in opposite directions and lose the fluidity that is interesting in the spring-like nature of the graph.
