## 1. Scraping Music Data from 7digital.com

We want to scrape data from 7digital.com because the website has short preview clips that can be downloaded. 

First, we want to set up the functions necessary to scrape the site. 

For each genre, we will get popular releases:
http://www.7digital.com/genre/alternative-indie/popular-releases?viewoption=list

And latest releases:
http://www.7digital.com/genre/alternative-indie/latest-releases?viewoption=list

For each album, we will get information for each song in the album:
http://www.7digital.com/artist/coldplay/release/parachutes-2

The information includes song name, duration, price, track number, and link to the preview mp3.

###1.1 Getting List of Genres

```python
#get list of genres and links
genres = [];

#open page with BS4
url = "http://www.7digital.com/genre/alternative"
request = urllib2.Request(url) 
page = urllib2.urlopen(request) 
soup2 = BeautifulSoup(page.read())

for link in soup2.select('a.genre-navigation-link'):
    genres.append(link['href'].split("/")[2])

print genres
```

###1.2 The Scraping Functions

```python
"""
Function
--------
get_albums

Given a genre name, return a list of dicts of album information

Inputs
------
genre_name : string
    The name of the genre

Returns
-------
genre_albums : list of dicts
    A list of dicts, each containing album information (name, artist, year) from popular releases and latest releases pages

Example
-------
>>> get_albums('alternative')
[{'artist': u'Lana Del Rey',
  'genre': 'alternative',
  'link': '/artist/lana-del-rey/release/born-to-die-3-1',
  'name': u'Born To Die',
  'year': u'2012'},
 {'artist': u'Coldplay',
  'genre': 'alternative',
  'link': '/artist/coldplay/release/parachutes-2',
  'name': u'Parachutes',
  'year': u'2006'},
  ...
"""    
def get_albums(genre_name):
    
    #init list
    genre_albums = []
    url2s = ["/popular-releases?viewoption=list","/latest-releases?viewoption=list"]
    
    for url2 in url2s:
        
        #open page with BS4
        url = "http://www.7digital.com/genre/"
        request = urllib2.Request(url+genre_name+url2) 
        page = urllib2.urlopen(request) 
        soup2 = BeautifulSoup(page.read())
        
        #get information about each album
        for alb in soup2.select('ul.list-layout p.unit-title'):
            
            #insert into dict for album
            alb_dict = {}
            alb_dict['name'] = alb.select('a.title')[0].text.strip()
            alb_dict['year'] = alb.select('span.small')[0].text.strip()[1:-1]
            alb_dict['artist'] = alb.select('a.artist')[0].text.strip()
            alb_dict['link'] = alb.select('a.title')[0]['href']
            alb_dict['genre'] = genre_name
            
            genre_albums.append(alb_dict)
            
    return genre_albums
```

```python
"""
Function
--------
get_songs

Given an album dict, return a dict of song dicts containing song information

Inputs
------
album : dict
    The album dict, one of those in the list created by the get_albums function

Returns
-------
alb_songs : dict of dicts
    A dict of song dicts, each containing song information (name, artist, year, album, genre, duration, price, track#, preview link). 
    The dict of songs in keyed by a unique song id generated from 7-digital's preview link

"""    
def get_songs(album):
    
    #init dict
    alb_songs = {}
        
    #open page with BS4
    url = "http://www.7digital.com/"
    request = (url+album['link']).encode('utf8')
    #request = urllib2.Request(uncoded_url) 
    
    #some pages have 404 errors. In that case, return empty dict. 
    try:
        page = urllib2.urlopen(request) 
    except (urllib2.HTTPError):
        return alb_songs
    soup2 = BeautifulSoup(page.read())
    
    #get information about each song
    for song_tr in soup2.findAll('tr',{'itemprop':'tracks'}):
        
        #get link for preview
        #if there is no tracklist preview, then we don't want the song
        try:
            link = song_tr.select("a.tracklist-preview")[0]['href']
        except (IndexError):
            continue
        
        #using preview link, generate a unique id to use as song key
        song_id = link.split("/")[2]+link.split("/")[4]+link.split("/")[6]
        
        #init empty dictionary
        song_dict = {}
        
        #insert into dict for album
        song_dict['album_name'] = album['name'] 
        song_dict['year'] = album['year']
        song_dict['artist'] = album['artist']
        song_dict['genre'] = album['genre']
        
        song_dict['name'] = song_tr.select("a.tracklist-preview")[0].text.strip()
        try:
            song_dict['duration'] = song_tr.select("td.tracktime")[0].text.strip()
        except:
            song_dict['duration'] = 'NaN'
        song_dict['preview_link'] = link
        song_dict['track_num'] = int(song_tr.select("td.tracknumber")[0].text.strip()[:-1])
        
        #some songs cannot be purchased separately (or other display errors), so put NaN           
        try:
            song_dict['price'] = float(song_tr.select("td.trackbuy span.price")[0].text.strip()[1:])
        except:
            song_dict['price'] = float('NaN')
        
        #if album has "various artists" find the real artist
        if (album['artist'] == "Various Artists"):
            try:
                song_dict['artist'] = song_tr.select("td.trackname span.trackadditional")[-1].text.strip().split("by ")[1]
            except (IndexError):
                song_dict['artist'] = 'NaN'
            
        #insert dict into big dict
        alb_songs[song_id] = song_dict
            
    return alb_songs
```

##3. Fast Fourier Transform (FFT)
FFT is an algorithm to perform a Discrete Fourier Transform, a method of signal processing that represents any signal as a supersposition of sine waves at different freqeuncies. We will use FFT to transform time-domain sound wave signals to the freqeuncy domain. We will then be able to represent each sound wave in numeric form as a vector of coefficients and phase angles. The coefficients describe the magnitude of the sine wave at each frequency and the phase angle determines the phase shift of the sine wave.

###3.2 Writing and Testing the FFT function

```python
#create custom errors for unusual sampling frequencies or preview mp3 lengths
class fs_error(Exception):
    pass
class len_error(Exception):
    pass
```

```python
"""
Function
--------
wav_fft

    Given a name of a wav file, convert the corresponding sound by FFT to the frequency domain. 

Inputs
------
wavfile : string
    The mp3 file name.

Returns
-------
(freq,outputSignal_a,outputSignal_m): tuple
    freq : np.array
        The list of frequencies of the sine waves used to decompose the signal. 
    outputSignal_a : np.array
        The list of phase angles that correspond to each sine wave phase shift. 
    outputSignal_m: np.array
        The list of coefficients that correspond to each signe wave magnitude. 
    
"""  
    
def wav_fft(wavfile):
    
    #take first 10 seconds on audio
    chunk = 10;
    
    #read in wav file
    (inputSignal, fs, bits) = audiolab.wavread("../../7digital_wav/"+wavfile+".wav")
    
    #if stereo, take first channel
    if (np.shape(np.shape(inputSignal))[0] == 2):
        inputSignal = inputSignal[:,0]
    
    #resample fs to 22050 because some songs have lower fs.
    if (fs != 22050):
        if (fs == 44100):
            #resample to 22050 Hz
            inputSignal = inputSignal[::2]
            fs = 22050
        else:
            #raise sampling frequency error
            raise fs_error(fs)
    
    #take first 10 seconds on audio
    inputSignal = inputSignal[:chunk*fs]
    
    #make sure the song is not shorter than sample time
    if len(inputSignal) != chunk*fs:
        raise len_error(len(inputSignal)/fs)
    
    #Perfrom Fast Fourrier Transform
    transformedSignal = np.fft.rfft(inputSignal)
    
    #take the angle and absolute value of the complex transformed signal
    outputSignal_a = [np.angle(x) for x in transformedSignal]
    outputSignal_m = [np.absolute(x) for x in transformedSignal]
    
    #generate corresponding frequency
    n= len(inputSignal) 
    val = float(fs)/n
    N = n//2 + 1
    results = np.arange(0, N, dtype=int)
    freq = results * val
    
    #return
    return (freq,outputSignal_a,outputSignal_m)
```

We create a helper function to convert a list of mp3 files and perform fft on them:

```python
"""
Function
--------
list_fft

    Given a list of the names mp3 files, convert all of them by FFT to the frequency domain. 

Inputs
------
song_list : list of strings
    The mp3 file names.

Returns
-------
songs_fft: list
    A list of lists. Each list contains the vector that is built by concatenating outputSignal_a and outputSignal_m 
    from the wav_fft function above. This way, no information is lost. 
    
"""  

def list_fft(song_list):
    
    #init storage array
    songs_fft = []
    
    for pid in song_list:
        
        #convert mp3 to wav
        convert_mp3(pid)
        
        try:
            #perform FFT, add to array
            x = wav_fft(pid)
            songs_fft.append(np.concatenate((x[1],x[2])))
            
        except fs_error as e:
            #notify if there is a frequency sampling error
            print "fs error:", e
            
        except len_error as e:
            #notify if there is a too short song error
            print "length error:", e
            
        #remove the wav to save storage space
        remove_wav(pid)
        
    return songs_fft
```

Note that the above function concatenates the angle data and the magnitude data from FFT, so no information is lost.

```python
"""
Function
--------
off_diag

        Helper-function - Given a matrix, create a list of one off-diganonal triangle. 

Inputs
------
matrix : list of lists
    A square matrix of numbers. 

Returns
-------
total: a np array
    A numpy array of the off-diganonal elements on one side of the diagonal. 
    
"""  
def off_diag(matrix):
    
    total = [];
    side = len(matrix)
    
    #make sure it's a square matrix
    assert side == len(matrix[0])
    
    for i in xrange(0,side):
        total += matrix[i][(i+1):].tolist()
        
    return np.array(total)
```

##5. Support Vector Machine: Determining the most well-defined categorizations
To identify the pairs of categories that have the most well-defined boudaries between them, we used a linear support vector machine to find hyperplane that separates two potential clusters on the PCA plane. Two clusters that can be separated well with a linear hyperplane represent two categories (i.e. two genres or two artists) that are meaningfully distinct. Conversely, if songs of two categories cannot be separated well, then those two categories may contain very similar-sounding songs. 

###5.1 Determining the most well-defined genres

```python
"""
Function
--------
svm_fft

        Perform a linear support vector machine on the PCA transformed fft data to find the linear 
        hyperplane separating two groups in a category. 

Inputs
------
cat1 : string
    The first group that will be analyzed (ex. pop)
    
cat2 : string
    The second group that will be analyzed (ex. rock)
    
category : string
    The category from which the groups will be drawn. (ex. genre)
    
plot : bool
    Whether the function should display the SVM plot
    
Returns
------
error : float
    Fraction of incorrect predictions in each group, averaged over both groups
    
"""  
def svm_fft(cat1,cat2,category, plot):
    #make a mask
    mask1 = sm_song_df[category].values == cat1
    mask2 = sm_song_df[category].values == cat2
    
    mask1_fft = [i for i,m in zip(fft_songs_filtered,mask1) if m]
    mask2_fft = [i for i,m in zip(fft_songs_filtered,mask2) if m]
    
    together_fft = mask1_fft + mask2_fft
    categs = [0]*len(mask1_fft) + [1]*len(mask2_fft)
    
    #SVM
    svm_X = pca.transform(together_fft)[:,:2]
    svm_Y = np.array(categs)
    
    #scale data
    svm_X = preprocessing.scale(svm_X)
    
    # fit the model with a linear SVM
    clf = svm.SVC(kernel='linear')
    clf.fit(svm_X, svm_Y)
    
    #create a predictor
    Y_hat = clf.predict(svm_X)
    
    #define error as the percentage of incorrect predictions in each group, averaged over both groups
    error1 = 1-sum((Y_hat==svm_Y)[svm_Y==1])/float(svm_Y.sum())
    error2 = 1-sum((Y_hat==svm_Y)[svm_Y==0])/float(len(svm_Y)-svm_Y.sum())
    error = (error1 + error2)/2.
    
    if plot:
        plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80,
                   facecolors='none', zorder=10)
        plt.scatter(svm_X[:, 0], svm_X[:, 1], c=svm_Y, zorder=10, cmap=plt.cm.Paired)
        
        #plt.axis('tight')
        x_min = -3
        x_max = 3
        y_min = -3
        y_max = 3
        
        XX, YY = np.mgrid[x_min:x_max:400j, y_min:y_max:400j]
        Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])
        
        # Put the result into a color plot
        Z = Z.reshape(XX.shape)
        plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
        plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                   levels=[-.5, 0, .5])
        
        plt.xlim(x_min, x_max)
        plt.ylim(y_min, y_max)
        
        plt.xticks(())
        plt.yticks(())
        
        plt.xlabel('Normalized PC 1')
        plt.ylabel('Normalized PC 2')
        plt.title('SVM on Songs in the '+category+"s "+ cat1+" and " +cat2+"\n error: "+str(round(error*100,2))+"%")
    
    return error
    
```

```python
#plot characteristic SVM results
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
svm_fft("pop","folk","genre", True)
plt.subplot(1,3,2)
svm_fft("electronica","folk","genre",True)
plt.subplot(1,3,3)
svm_fft("folk","punk","genre",True)
plt.show()
```

There appears to be a well-defined distinction between songs classified as pop and those classifed as folk. These distinctions make intuitive sense from experience.

###6.4 SVM on PCA data

```python
"""
Function
--------
svm_fft2

        Perform a linear support vector machine on the PCA 2D plane to find the linear 
        hyperplane separating two groups in a category. 

Inputs
------
cat1 : string
    The first group that will be analyzed (ex. pop)
    
cat2 : string
    The second group that will be analyzed (ex. rock)
    
category : string
    The category from which the groups will be drawn. (ex. genre)
    
plot : bool
    Whether the function should display the SVM plot
    
df : pandas DataFrame
    Data Frame form which song data will be pulled
    
X : list of lists
    2D PCA data
    
X_index : list of strings
    The song_df indices that correspond to the X PCA data
    
Returns
------
error : float
    Fraction of incorrect predictions in each group, averaged over both groups
    
"""  
def svm_fft2(cat1,cat2,category,plot,df,X, X_index):
    #make a mask
    mask1 = df[category].values == cat1
    mask2 = df[category].values == cat2
    
    #find index for ech mask
    index1 = df[mask1].index.values
    index2 = df[mask2].index.values
    
    #get PCA data for each filter
    mask1_fft = [x for x,i in zip(X,X_index) if (int(i) in index1)]
    mask2_fft = [x for x,i in zip(X,X_index) if (int(i) in index2)]
    
    #create X and Y vectors for svm
    svm_X = mask1_fft + mask2_fft
    svm_Y = np.array([0]*len(mask1_fft) + [1]*len(mask2_fft))
    
    #scale data
    svm_X = preprocessing.scale(svm_X)
    
    # fit the model with a linear SVM
    clf = svm.SVC(kernel='linear')
    clf.fit(svm_X, svm_Y)
    
    #create a predictor
    Y_hat = clf.predict(svm_X)
    
    #define error as the percentage of incorrect predictions in each group, averaged over both groups
    error1 = 1-sum((Y_hat==svm_Y)[svm_Y==1])/float(svm_Y.sum())
    error2 = 1-sum((Y_hat==svm_Y)[svm_Y==0])/float(len(svm_Y)-svm_Y.sum())
    error = (error1 + error2)/2.
    
    if plot:
        plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80,
                   facecolors='none', zorder=10)
        plt.scatter(svm_X[:, 0], svm_X[:, 1], c=svm_Y, zorder=10, cmap=plt.cm.Paired)
        
        #plt.axis('tight')
        x_min = -3
        x_max = 3
        y_min = -3
        y_max = 3
        
        XX, YY = np.mgrid[x_min:x_max:400j, y_min:y_max:400j]
        Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])
        
        # Put the result into a color plot
        Z = Z.reshape(XX.shape)
        plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
        plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                   levels=[-.5, 0, .5])
        
        plt.xlim(x_min, x_max)
        plt.ylim(y_min, y_max)
        
        plt.xticks(())
        plt.yticks(())
        
        plt.xlabel('Normalized Principal Component 1')
        plt.ylabel('Normalized Principal Component 2')
        plt.title('SVM on Songs in the '+category+"s "+ cat1+" and " +cat2+"\n error: "+str(round(error*100,2))+"%")
    
    return error
```

```python
def svm_matrix(df,category,thresh,verbose=False):
    
    #get unique genres
    genres = svm_df[category].unique()
    
    #use only genres with more than thresh songs
    big_genres = []
    for eachg in genres:
        if len(df[df[category]==eachg]) >= thresh:
            big_genres.append(eachg)
            
    matrix_data = []
    
    #iterate over a matrix of genres to get each pair
    for e,i in enumerate(big_genres):
        
        if verbose:
            print e,"of", len(big_genres)
        
        for j in big_genres[(e+1):]:
            tup = [svm_fft2(i,j,category,False,df,X_1000, index_1000)]+[i,j]
            matrix_data.append(tup)
            
    return matrix_data
```

###7.1 Building the model on the training set

```python
def knn(X_knn_train, X_knn_test, index_ts,index_knn_test, X_sd, df, k, category, plot=False):
    X_knn = X_knn_train/X_sd
    
    #get category groups list
    groups = df[category].unique().tolist()
    
    #create X and Y for KNN
    Y_knn = []
    for ind in index_ts:
        group = df.loc[[int(ind)]][category].values[0]
        #assign numerical representations to each group
        Y_knn.append(groups.index(group))
    
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(k, weights='uniform')
    clf.fit(X_knn, Y_knn)
        
    if plot:
        h = .02  # step size in the mesh
        
        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, m_max]x[y_min, y_max].
        x_min, x_max = X_knn[:, 0].min() - 1, X_knn[:, 0].max() + 1
        y_min, y_max = X_knn[:, 1].min() - 1, X_knn[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        
        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        plt.figure()
        plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
        
        # Plot also the training points
        plt.scatter(X_knn[:, 0], X_knn[:, 1], c=Y_knn, cmap=plt.cm.Paired)
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.title("%i-Class classification (k = %i)" % (len(groups),radius))
        
        plt.show()
    
    #get ground truth for test set
    Y_knn_test = []
    for ind in index_knn_test:
        group = df.loc[[int(ind)]][category].values[0]
        #assign numerical representations to each group
        Y_knn_test.append(groups.index(group))
    
    #return error aspercentage of correct IDs
    return 1- sum([float(i) for i in ((clf.predict(X_knn_test/X_sd)- Y_knn_test)==0).tolist()])/len(Y_knn_test)
```

We colored the Normalized PCA plane accoring to the KNN model with an arbitrary k=12 to visualize how, accoridng to the current model, points in the space would be classified as different genres. Importantly, we did not group, merge, or exclude any genres from analysis. Thus, there appears to be a large mix of colors.

In this section, we vary the value of k in the KNN model to see how the prediction error changes. We use the test set to measure error. Note that the PCa was not fitted with the test set, and the "normalization" was done with a the standard deviation from the training set. Thus, the test set is truly a new set of data that has not affected the initial creation of the model in any way.
